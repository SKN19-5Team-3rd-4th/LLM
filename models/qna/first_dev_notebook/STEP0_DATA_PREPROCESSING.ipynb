```python
import json
DATA_PATH = "../../datas/post.json"
with open(DATA_PATH, "r", encoding="utf-8") as f:
    raw_data = json.load(f)
```

```python
raw_data[0]
```

```python
import pandas as pd 

df = pd.DataFrame(raw_data)[['channel', 'author', 'content', 'comments', 'post_url']]
df = df.drop_duplicates(['content'])
df = df[(df['channel']=='식물Q&A') & (df['author']!='그로로알림이')]

list_cols = [col for col in df.columns if df[col].apply(lambda x: isinstance(x, list)).any()]

for col in list_cols:
    df = df.explode(col)
df = df.reset_index(drop=True)

df['comments'] = df['comments'].apply(lambda x: x if not 'content' in str(x) else x['content'])
df['post_id'] = df['post_url'].apply(lambda x: x.split('/')[-1])
df = df.drop(['channel', 'author', 'post_url'], axis=1)

# max(df['content'].apply(lambda x: len(x)))
# df[df['content'].apply(lambda x: len(x)==514)].to_csv('./max_qna.csv', encoding='utf-8-sig')

df.to_csv('../../cache/comments.csv', index=False, encoding='utf-8-sig')
```

```python
df_content = df.drop_duplicates('post_id').reset_index(drop=True)
df_content.info()
```

```python
df_comment = df.groupby('post_id')['comments'].agg(lambda x: ','.join(x.astype(str))).reset_index()
df_comment.info()
```

```python
df_merge = pd.merge(df_content, df_comment, on='post_id', how='inner')
df_merge.info()
```

```python
final_df = df_merge[['post_id','content','comments_y']]
final_df.info()
```

```python
# {
#     "id": ,
#     "question": ,
#     "answer": ,
#     "metadata": {
#         "post_id":  
#         "source_platform": "groro"
#     }
# }

# 응답시 >> (필요한정보) + 다른사람들도 비슷한걸 궁금해 했고(공감), 이렇게 도움을 받기도했어(참고정보) + 전문가의 의견은 아니므로 참고만 해야해!
```

```python
final_df.tail()
```

```python
data = []
for idx, row in final_df.iterrows():
    q = row['content']
    a = row['comments_y']
    post_id = row['post_id']
    
    data.append({
        "ids": "groro" + "_" + str(idx + 1) + "_" + str(post_id),
        "question": q,
        "answer": a,
        "metadata": {
            "post_id": str(post_id),
            "source_platform": "groro"
        }
    })
```
